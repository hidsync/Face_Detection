```python
#FACE RECOGNITION
import face_recognition
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import cv2
%matplotlib inline
```

```python
image=cv2.imread('bill2.jpg')
image=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
face_locations=face_recognition.face_locations(image)
no_of_faces=len(face_locations)
print("Found {} faces".format(no_of_faces))
```

```python
plt.imshow(image)
ax=plt.gca()
for face_location in face_locations:
    top,right,bottom,left=face_location
    x,y,w,h=left,top,right,bottom
    print("T{}, L{}, B{}, R{}".format(x,y,w,h))
    rect=Rectangle((x,y),w-x,h-y,fill=False,color='red')
    ax.add_patch(rect)
plt.show()
```

```python
#Face Recognition
import face_recognition
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
import numpy as np
import cv2
%matplotlib inline
```

```python
image=cv2.imread('bill.jpg')
bill=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image=cv2.imread('warren.jpg')
warren=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image=cv2.imread('diana.jpg')
diana=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image=cv2.imread('oprah.jpg')
oprah=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image=cv2.imread('tomh.jpg')
tom=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

image=cv2.imread('ed.jpg')
ed=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
bill_c=face_recognition.face_encodings(bill)[0]
warren_c=face_recognition.face_encodings(warren)[0]
diana_c=face_recognition.face_encodings(diana)[0]
oprah_c=face_recognition.face_encodings(oprah)[0]
tom_c=face_recognition.face_encodings(tom)[0]
ed_c=face_recognition.face_encodings(ed)[0]
known_face=[bill_c, warren_c, diana_c, oprah_c, tom_c, ed_c]
```

```python
image=cv2.imread('pic3.jpg')
new_im=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(new_im)
new_c=face_recognition.face_encodings(new_im)
```

```python
from scipy.spatial import distance
for new_cs in new_c:
    results=[]
    for known in known_face:
        d=distance.euclidean(new_cs,known)
        results.append(d)
    threshold=0.6
    results=np.array(results) <= threshold
    name="Unknown"
    if results[0]:
        name="Bill Gates"
    elif results[1]:
        name="Warren Buffet"
    elif results[2]:
        name="Princess Diana"
    elif results[3]:
        name="Oprah Winfrey"
    elif results[4]:
        name="Tom Hanks"
    elif results[5]:
        name="Ed Sheeran"
    
    print(f"Found {name} in the photo.")
```
